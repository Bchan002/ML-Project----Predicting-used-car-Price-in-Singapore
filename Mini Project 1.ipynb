```python
#drop NA values 
sg_clean.dropna(inplace=True)
sg_clean.info() 
#We have now 17 features in our columns
```

```python
#Removing whitespace from Vehicle Type 
sg_clean['VEHICLE_TYPE'].apply(str.strip)
sg_clean['VEHICLE_TYPE'].apply(str.lstrip)

sg_clean
```

```python
#Removing whitespace from Brand 
sg_clean['BRAND'].apply(str.strip)

sg_clean
```

```python
#Renaming Brands of Cars 
sg_clean.loc[sg_clean['BRAND'] == 'Aston','BRAND'] = 'Aston Martin'
sg_clean.loc[sg_clean['BRAND'] == 'Land','BRAND'] = 'Land Rover'
sg_clean.loc[sg_clean['BRAND'] == 'Alfa', 'BRAND'] = 'Alfa Romeo'

sg_clean 
```

## Part 2.1 Car Brand Categorization 
There are 7 classes for Vehicle Type, namely Mid-Sized Sedan, Luxury Sedan, Hatchback, MPV, SUV, Sports Car, Stationwagon. Each vehicle is assigned only one Vehicle Type. This classification can be vague because premium and mass-market brands produce vehicles in all of these segments. For instance, a Toyota Camry, a BMW and a Rolls-Royce Ghost are all classified under Luxury Sedan as per sgcarmart’s definition. Vehicle Type is a poor way to segment the vehicles. 

Instead use Brand as the categorization instead as it tells alot on the level of prestige and perceived desirability of a vehicle. Example, hear someone driving a Mercedes, first thought is this car is prestigous' Thus make sense to categorize the Vehicles into their respective brands 

Cars are split into 6 segments are Exotic, Ultra Luxury, Luxury, Mid Level, Economy and Budget.

The proposed segments and their respective makes are as follows:

Exotic: Aston Martin, Ferrari, Lamborghini, McLaren

Ultra Luxury: Bentley, Land Rover, Maserati, Porsche, Rolls-Royce

Luxury: Audi, BMW, Jaguar, Jeep, Lexus, Lotus, Mercedes-Benz, Mitsuoka, Volvo

Mid Level: Alfa Romeo, Chrysler, Infiniti, MINI, Opel, Saab, Volkswagen

Economy: Chevrolet, Citroen, Fiat, Ford, Honda, Hyundai, Kia, Mazda, Mitsubishi, Nissan, Peugeot, Renault, Skoda, Ssangyong, 

Subaru, Suzuki, Toyota

Budget: Daihatsu, Proton
